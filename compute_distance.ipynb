`compute_distance` find the distance to the closest neighboring gene of a putative lncRNA transcript.

```python
import pandas as pd
import pprint
```

```python
gtf_file = 'stringtie_merged.gtf'
master_excel = 'Master_Sheet_20170719.xlsx'
```

```python
lincs = pd.read_excel(master_excel, sheetname='putative lincRNA')
genes = pd.read_excel(master_excel, sheetname='mRNA')
```

```python
# Read in the GTF annotation
annotation = pd.read_table(gtf_file, header=None, comment='#')
```

```python
chromosomes = list()
start_coords = list()
stop_coords = list()
for locus in genes.loc[:, 'locus']:
    chromosome, coords = locus.split(':')
    chromosomes.append(chromosome)
    start, stop = coords.split('-')
    start_coords.append(start)
    stop_coords.append(stop)
genes = genes.join(pd.DataFrame({'chromosome': chromosomes, 'start': start_coords, 'stop': stop_coords}))
```

```python
chromosomes = list()
start_coords = list()
stop_coords = list()
for locus in lincs.loc[:, 'locus']:
    chromosome, coords = locus.split(':')
    chromosomes.append(chromosome)
    start, stop = coords.split('-')
    start_coords.append(start)
    stop_coords.append(stop)
lincs = lincs.join(pd.DataFrame({'chromosome': chromosomes, 'start': start_coords, 'stop': stop_coords}))
```

```python
genes
```

```python
lincs
```

```python
# Rename the columns in the GTF annotation dataframe
annotation = annotation.rename(index=str, columns={0: 'chromosome', 1: 'source', 2: 'feature', 3: 'start', 4: 'stop', 5: 'score', 6: 'strand', 7: 'frame', 8: 'attribute'})
```

```python
annotation
```

```python
# We only want transcript features
transcripts = annotation.loc[annotation.feature == 'transcript', :]
```

```python
transcripts
```

```python
# parse the attributes of the transcripts
gene_ids = list()
transcript_ids = list()
gene_names = list()
ref_gene_ids = list()
for attribute in transcripts.loc[:, 'attribute']:
    attrs = attribute.split(';')
    attrs = [field.strip(' ') for field in attrs]
    gene_ids.append(attrs[0].split(' ')[1].strip('\"'))
    transcript_ids.append(attrs[1].split(' ')[1].strip('\"'))
    # >3 because the split yields an extra null field due to the trailing ';'
    if len(attrs) > 3:
        gene_names.append(attrs[2].split(' ')[1].strip('\"'))
        ref_gene_ids.append(attrs[3].split(' ')[1].strip('\"'))
    else:
        gene_names.append('-')
        ref_gene_ids.append('-')

# Here, we must first re-set the index so that we can join the two data frames together
# When we re-index we do it in place and we don't save it as a new column
transcripts.reset_index(drop=True, inplace=True)
transcripts = transcripts.join(pd.DataFrame({'gene_id': gene_ids, 'transcript_id': transcript_ids, 'gene_name': gene_names, 'ref_gene_id': ref_gene_ids}))
```

```python
transcripts
```

```python
# We want all the transcripts that are in the mRNA master sheet
mRNA_BED = transcripts.loc[transcripts.ref_gene_id.isin(genes.gene), ['chromosome', 'start', 'stop', 'ref_gene_id']]
```

```python
linc_BED = lincs.loc[:, ['chromosome', 'start', 'stop', 'major_iso_id']]
```

```python
# We should sort the lincs by chromosome and then by start
# Notice the start column is being treated as a string, giving us the incorrect sorting
linc_BED.sort_values(['chromosome', 'start'])
```

```python
# We can convert columns to numeric by making use of pd.to_numeric and apply that
# to the entire data frame. Setting errors='ignore' tells it to just leave the columns
# that can't be converted alone. We pass that output to sort_values.

# Now we see the numerically sorted output
linc_BED.apply(pd.to_numeric, errors='ignore').sort_values(['chromosome', 'start'])
```

```python
linc_BED = linc_BED.apply(pd.to_numeric, errors='ignore').sort_values(['chromosome', 'start'])
```

```python
# Now do the same for mRNA_BED
mRNA_BED.apply(pd.to_numeric, errors='ignore').sort_values(['chromosome', 'start'])
```

```python
mRNA_BED = mRNA_BED.apply(pd.to_numeric, errors='ignore').sort_values(['chromosome', 'start'])
```

```python
# Now we'll write them out to file and go run bedtools closest
mRNA_BED.to_csv('mRNA.bed', sep='\t', index=False, header=False)
linc_BED.to_csv('lincRNA.bed', sep='\t', index=False, header=False)
```
