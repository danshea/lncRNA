`make_FPKM_files` will create files for FPKM analysis using bedtools. That data will be plotted using `circos`.

```python
# File with transcript information
transcript_file = 'transcripts.tsv'
# File with total genes seen
genes_file = 'Total_genes.xlsx'
# File with total incRNAs and NATs
lnc_file = 'putative_lncrnas.xlsx'
# File with total lincRNAs and unannotated genes
linc_file = 'putative_lincrnas.xlsx'
```

```python
import pandas as pd
```

```python
# Read in the data from various sources
transcripts   = pd.read_table(transcript_file)
mRNA          = pd.read_excel(genes_file)
incRNA        = pd.read_excel(lnc_file, sheetname='Intronic')
NAT           = pd.read_excel(lnc_file, sheetname='NAT')
lincRNA       = pd.read_excel(linc_file, sheetname='Intergenic')
p_mRNA        = pd.read_excel(linc_file, sheetname='Unannotated genes')
```

```python
# We only want entries where status == "OK" in the data tables
mRNA    = mRNA[mRNA.status == "OK"]
incRNA  = incRNA[incRNA.status == "OK"]
NAT     = NAT[NAT.status == "OK"]
lincRNA = lincRNA[lincRNA.status == "OK"]
p_mRNA  = p_mRNA[p_mRNA.status == "OK"]
```

```python
mRNA_transcripts = pd.merge(mRNA, transcripts, left_on='major_iso_id', right_on='transcript_id')
```

```python
p_mRNA_transcripts = pd.merge(p_mRNA, transcripts, left_on='major_iso_id', right_on='transcript_id')
```

```python
incRNA_transcripts = pd.merge(incRNA, transcripts, left_on='major_iso_id', right_on='transcript_id')
```

```python
lincRNA_transcripts = pd.merge(lincRNA, transcripts, left_on='major_iso_id', right_on='transcript_id')
```

```python
NAT_transcripts = pd.merge(NAT, transcripts, left_on='major_iso_id', right_on='transcript_id')
```

```python
# NV Condition FPKM (all STATUS="OK")
mRNA_NV    = mRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_1']]
p_mRNA_NV  = p_mRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_1']]
incRNA_NV  = incRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_1']]
lincRNA_NV = lincRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_1']]
NAT_NV     = NAT_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_1']]
# V Condition FPKM (all STATUS="OK")
mRNA_V    = mRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_2']]
p_mRNA_V  = p_mRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_2']]
incRNA_V  = incRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_2']]
lincRNA_V = lincRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_2']]
NAT_V     = NAT_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'value_2']]
# log2(fold_change) (all STATUS="OK")
mRNA_fc    = mRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
p_mRNA_fc  = p_mRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
incRNA_fc  = incRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
lincRNA_fc = lincRNA_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
NAT_fc     = NAT_transcripts.loc[:, ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
# log2(fold_change) (all STATUS="OK" and significant="yes")
mRNA_DE_fc    = mRNA_transcripts.loc[mRNA_transcripts.significant == 'yes', ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
p_mRNA_DE_fc  = p_mRNA_transcripts.loc[p_mRNA_transcripts.significant == 'yes', ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
incRNA_DE_fc  = incRNA_transcripts.loc[incRNA_transcripts.significant == 'yes', ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
lincRNA_DE_fc = lincRNA_transcripts.loc[lincRNA_transcripts.significant == 'yes', ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
NAT_DE_fc     = NAT_transcripts.loc[NAT_transcripts.significant == 'yes', ['chromosome', 'start', 'stop', 'transcript_id', 'log2(fold_change)']]
```

```python
# Combine all of the DE lncRNA fold-change data
lncRNA_DE_fc = pd.concat(objs=[incRNA_DE_fc, lincRNA_DE_fc, NAT_DE_fc])
```

```python
# Sanity check that the counts match (should be 549)
lncRNA_DE_fc.shape
```

```python
# Sort the output by chromosome and start position
# NV Condition FPKM (all STATUS="OK")
mRNA_NV    = mRNA_NV.sort_values(['chromosome', 'start'])
p_mRNA_NV  = p_mRNA_NV.sort_values(['chromosome', 'start'])
incRNA_NV  = incRNA_NV.sort_values(['chromosome', 'start'])
lincRNA_NV = lincRNA_NV.sort_values(['chromosome', 'start'])
NAT_NV     = NAT_NV.sort_values(['chromosome', 'start'])
# V Condition FPKM (all STATUS="OK")
mRNA_V    = mRNA_V.sort_values(['chromosome', 'start'])
p_mRNA_V  = p_mRNA_V.sort_values(['chromosome', 'start'])
incRNA_V  = incRNA_V.sort_values(['chromosome', 'start'])
lincRNA_V = lincRNA_V.sort_values(['chromosome', 'start'])
NAT_V     = NAT_V.sort_values(['chromosome', 'start'])
# log2(fold_change) (all STATUS="OK")
mRNA_fc    = mRNA_fc.sort_values(['chromosome', 'start'])
p_mRNA_fc  = p_mRNA_fc.sort_values(['chromosome', 'start'])
incRNA_fc  = incRNA_fc.sort_values(['chromosome', 'start'])
lincRNA_fc = lincRNA_fc.sort_values(['chromosome', 'start'])
NAT_fc     = NAT_fc.sort_values(['chromosome', 'start'])
# log2(fold_change) (all STATUS="OK" and significant="yes")
mRNA_DE_fc    = mRNA_DE_fc.sort_values(['chromosome', 'start'])
p_mRNA_DE_fc  = p_mRNA_DE_fc.sort_values(['chromosome', 'start'])
incRNA_DE_fc  = incRNA_DE_fc.sort_values(['chromosome', 'start'])
lincRNA_DE_fc = lincRNA_DE_fc.sort_values(['chromosome', 'start'])
NAT_DE_fc     = NAT_DE_fc.sort_values(['chromosome', 'start'])
# lncRNA log2(fold_change) (all STATUS="OK" and significant="yes")
lncRNA_DE_fc  = lncRNA_DE_fc.sort_values(['chromosome', 'start'])
```

```python
# Write out NV FPKM tables
mRNA_NV.to_csv('mRNA_NV_FPKM.tsv', sep='\t', header=False, index=False)
p_mRNA_NV.to_csv('p_mRNA_NV_FPKM.tsv', sep='\t', header=False, index=False)
incRNA_NV.to_csv('incRNA_NV_FPKM.tsv', sep='\t', header=False, index=False)
lincRNA_NV.to_csv('lincRNA_NV_FPKM.tsv', sep='\t', header=False, index=False)
NAT_NV.to_csv('NAT_NV_FPKM.tsv', sep='\t', header=False, index=False)
```

```python
# Write out V FPKM tables
mRNA_V.to_csv('mRNA_V_FPKM.tsv', sep='\t', header=False, index=False)
p_mRNA_V.to_csv('p_mRNA_V_FPKM.tsv', sep='\t', header=False, index=False)
incRNA_V.to_csv('incRNA_V_FPKM.tsv', sep='\t', header=False, index=False)
lincRNA_V.to_csv('lincRNA_V_FPKM.tsv', sep='\t', header=False, index=False)
NAT_V.to_csv('NAT_V_FPKM.tsv', sep='\t', header=False, index=False)
```

```python
# Write out total fold-change tables
mRNA_fc.to_csv('mRNA_total_fc_FPKM.tsv', sep='\t', header=False, index=False)
p_mRNA_fc.to_csv('p_mRNA_total_fc_FPKM.tsv', sep='\t', header=False, index=False)
incRNA_fc.to_csv('incRNA_total_fc_FPKM.tsv', sep='\t', header=False, index=False)
lincRNA_fc.to_csv('lincRNA_total_fc_FPKM.tsv', sep='\t', header=False, index=False)
NAT_fc.to_csv('NAT_total_fc_FPKM.tsv', sep='\t', header=False, index=False)
```

```python
# Write out DE fold-change tables
mRNA_DE_fc.to_csv('mRNA_fc_FPKM.tsv', sep='\t', header=False, index=False)
p_mRNA_DE_fc.to_csv('p_mRNA_fc_FPKM.tsv', sep='\t', header=False, index=False)
incRNA_DE_fc.to_csv('incRNA_fc_FPKM.tsv', sep='\t', header=False, index=False)
lincRNA_DE_fc.to_csv('lincRNA_fc_FPKM.tsv', sep='\t', header=False, index=False)
NAT_DE_fc.to_csv('NAT_fc_FPKM.tsv', sep='\t', header=False, index=False)
```

```python
# Write out fold-change table for total lncRNA
lncRNA_DE_fc.to_csv('lncRNA_fc_FPKM.tsv', sep='\t', header=False, index=False)
```
